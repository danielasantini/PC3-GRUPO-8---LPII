```python
import requests
response=requests.get('https://api.escuelajs.co/api/v1/products')

data2=response.text
```

```python
import json

info = json.loads(data2)
info
#print('número de usuarios:', len(info))
```

```python
id=[]
titulos=[]
precios=[]
descripcion=[]
categoria=[]

for item in info:
  id.append(item['id']-1)
  titulos.append(item['title'])
  precios.append(item['price'])
  descripcion.append(item['description'])
  categoria.append(item['category']['name'])
  #print('Attribute', item['x'])
```

Convertir a tabla

```python
import pandas as pd

Tabla = pd.DataFrame(zip(titulos,precios,descripcion,categoria),
                        columns=['Artículo','Precios','Descripcion','Categoria']
                        )
Tabla
```

Algunos estadísticos

Categorias más compradas

```python
def unique_values(lista):
    unicos = []
    for valor in lista:
      if valor not in unicos:
        unicos.append(valor)
    return unicos
```
